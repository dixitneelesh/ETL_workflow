```python
import requests
import pandas as pd
from sqlalchemy import create_engine

# Step 1: Extract Data from API
url = "https://jsonplaceholder.typicode.com/users"

response = requests.get(url)

data = response.json()

# Step 2: Flatten Nested JSON
df = pd.json_normalize(data)

# View columns
print(df.columns)
```

```python
# Step 3: SQL Server Connection
engine = create_engine(
    'mssql+pyodbc://LAPTOP-57BAC2EN\\SQLEXPRESS/master?driver=ODBC+Driver+17+for+SQL+Server&trusted_connection=yes'
)

# Step 4: Load Data into SQL Server
df.to_sql(
    name='users_data',
    con=engine,
    if_exists='replace',
    index=False
)

print("Data Loaded Successfully")
```
